```python
import pandas as pd

# Đọc file dữ liệu hoàn chỉnh (đã tạo ở Task 4)
df = pd.read_csv('complete_data.csv')

print("\n--- Kiểm tra dữ liệu Task 5 ---")
print(df.head())

# 1. Pivot: Tổng lượng điện tiêu thụ theo house_type × tariff_level
pivot_consumption = pd.pivot_table(
    df,
    values='consumption_kwh',
    index='house_type',
    columns='tariff_level',
    aggfunc='sum'
)

print("\n--- Pivot: Tổng điện tiêu thụ theo Loại nhà × Bậc thang ---")
print(pivot_consumption)

# 2. Pivot: Tiền điện trung bình theo region × house_type
pivot_cost = pd.pivot_table(
    df,
    values='estimated_cost',
    index='region',
    columns='house_type',
    aggfunc='mean'
).round(0)

print("\n--- Pivot: Tiền điện trung bình theo Khu vực × Loại nhà ---")
print(pivot_cost)

```
